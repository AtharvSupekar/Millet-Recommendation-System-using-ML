```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Millet Survey data.csv")
```

```python
df.tail(1)
```

```python
data = df.drop(['How often do you exercise and what is the duration of your workout?','You need millet for following benifts','Are you interested in learning more about the nutritional benefits of millet and incorporating it into a balanced diet or your sports nutrition plan?','Do you have any unique millet based recipes or nutrition tips for other active women like you?',"Any additional comments or suggestions related to millet, sports nutrition and women's health you would like to suggest."],axis=1)
```

```python
data.head(1)
```

```python
data1 = data.rename(columns={'What type of sports or workouts do you regularly engage in?': 'Sports Activity', 'Do you follow any post-workout nutrition?': 'Post-workout Nutrition','How often do you incorporate millet into your meals or regular diet?': 'Millet in diet','Specify any problem you face during the sports activity ( eg. ligament issue, knee pain, skin related issue etc.)':'Problems during sports','Are you aware of the nutritional benefits of millet?':'Awareness','Do you include millet in your diet as a part of your sports nutrition plan?':'Millet in Nutrition Plan'})
```

```python
data1.head(1)
```

```python
data1 = data1.rename(columns = {'If yes, please specify the types and amount of food you consume after workout. (For eg. protein sources, carbohydrates, fiber-rich food..)':'Food Post Workout'})
```

```python
data1.head(1)
```

```python
data2 = data1.drop(columns = [df.columns[6]],axis = 1)
```

```python
data2.head(1)
```

```python
data2
```

### Percentage of people who follow post workout nutrition

```python
# Count the number of people who follow and don't follow a post-workout nutrition plan
post_workout_counts = data2['Post-workout Nutrition'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(post_workout_counts, labels=post_workout_counts.index, autopct='%1.1f%%', startangle=180)
plt.title('Percentage of People Following Post-Workout Nutrition')
plt.legend()
plt.axis('equal')

# Show the plot
plt.show()
```

### Awareness about millets age wise

```python
awareness_counts_by_age_group = data2.groupby('Age Group')['Awareness'].value_counts().unstack().fillna(0)

# Plotting
plt.figure(figsize=(10, 6))
awareness_counts_by_age_group.plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'])
plt.title('Awareness about Millets by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Awareness', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
awareness_counts_by_age_group = data2.groupby('Age Group')['Millet in Nutrition Plan'].value_counts().unstack().fillna(0)

# Plotting
plt.figure(figsize=(10, 6))
awareness_counts_by_age_group.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.title('Inclusion of Millets in Diet by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Awareness', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
problems_by_age_group = data2.groupby('Age Group')['Problems faced during Workout'].value_counts().unstack().fillna(0)
# Plotting
plt.figure(figsize=(10, 6))
problems_by_age_group.plot(kind='bar', stacked=True)
plt.title('Problems Faced During Workout by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Problem Faced', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
